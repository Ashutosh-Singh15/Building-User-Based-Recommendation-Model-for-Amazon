# Recommendation model

```python
#Reshaping movies
movies_reshaped=movies.melt(id_vars='user_id',value_vars=movies.columns[1:],var_name='movie',value_name='rating')
```

```python
movies_reshaped
```

```python
#filling nan values with zero(filling nan values with 0 removes bias) before applying SVD algorithm
movies_reshaped.fillna(0,inplace=True)
```

```python
movies_reshaped
```

# Model Building

```python
#Using Surprise library for model building
from surprise import Dataset
from surprise import Reader
from surprise import SVD #Using Singular value decompostion as the ml algorithm for user based prediction
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise.model_selection import GridSearchCV
```

```python
#Setting the rating scale
read=Reader(rating_scale=(-1,10))
```

```python
#Importing movies dataset into a surprise library dataset
surprise_data=Dataset.load_from_df(movies_reshaped[['user_id','movie','rating']],read)
surprise_data
```

```python
#Splitting data into train and test data
train_data,test_data=train_test_split(surprise_data,test_size=0.25,random_state=16)
```

```python
#creating and fitting svd object
svd=SVD()
svd.fit(train_data)
```

```python
#predicting ratings on test data
prediction=svd.test(test_data)
```

```python
#calculating rmse
accuracy.rmse(prediction)
```

```python
#Predicting specific ratings
svd.predict('A1IMQ9WMFYKWH5','Movie206',r_ui=5.0,verbose=True)
```

# Tuning Parameters for better performance

```python
#Finding better hyperparameters for svd
params={'n_epochs':[20,30],'lr_all':[0.002, 0.005],'reg_all':[0.4, 0.6],'n_factors':[50,60]}
from surprise.model_selection import GridSearchCV
gs=GridSearchCV(SVD,params,measures=['rmse','mae'],cv=3)
gs.fit(surprise_data)
```

```python
gs.best_params
```

```python
###Seemingly better hyperparameters:
#no. of iterations=20
#learning rate=0.002
#regularization parameter=0.4
#no. of matrix factors=60
```

```python
# Creating SVD object with seemingly better parameters
svd=SVD(n_epochs=20,lr_all=0.002,reg_all=0.4,n_factors=60)
```

```python
#fitting the data
svd.fit(train_data)
```

```python
#predicting rating on test set
predictions=svd.test(test_data)
```
